```python
data = [
    (1, '2021-01-01 10:00:00'),
    (1, '2021-01-01 11:00:00'),
    (1, '2021-01-01 12:00:00'),
    (2, '2021-01-01 12:00:00'),
    (2, '2021-01-01 13:00:00'),
    (2, '2021-01-01 14:00:00'),
]

columns = ["id", "datetime"]
df = spark.createDataFrame(data=data, schema=columns)
df.createOrReplaceTempView("window_test")
df.show()
```

```python
df.select(min("datetime").alias("min_datetime"), max("datetime").alias("max_datetime")).show()
```

```python
df_row = Window.orderBy("datetime")
df_rank = Window.partitionBy("id").orderBy("datetime")

df.withColumn("rank_number", rank().over(df_rank)).show()
df.withColumn("row_number", row_number().over(df_row)).show()
```

```python
data = [
    (1, '2021-01-01', 100, 'A'),
    (1, '2021-01-31', 105, 'A'),
    (2, '2021-02-04', 160, 'B'),
    (2, '2021-02-07', 145, 'B'),
]

columns = ["id", "date", "score", "type"]
df = spark.createDataFrame(data=data, schema=columns)
df.show()
```

```python
# 1. Definimos a janela: particionamos pelo ID e ordenamos pela DATA (decrescente para pegar o mais novo)
windowSpec = Window.partitionBy("id").orderBy(col("date").desc())

# 2. Aplicamos o row_number e filtramos apenas a primeira linha de cada grupo
df_deduplicado = df.withColumn("row_num", row_number().over(windowSpec)) \
                   .filter(col("row_num") == 1) \
                   .drop("row_num")

df_deduplicado.show()
```
